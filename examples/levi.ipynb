```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cs = [0.5, 1, 2, 4, 8] 
mu = 0          
x = np.linspace(mu, 5, 100) 

for c in cs:
    l = Levy(c, mu)
    dens = l.pdf(x)
    plt.plot(x, dens, label=('Плотность равпределения c c=' + str(c)))
    plt.title('График плотности распределения c различными параметрами c' )

plt.xlabel('x')
plt.ylabel('pdf(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
cs = [0.5, 1, 2, 4, 8] 
mu = 0          
x = np.linspace(mu, 5, 100) 
x = x[x!=0]

for c in cs:
    l = Levy(c, mu)
    dens = l.cdf(x)
    plt.plot(x, dens, label=('Функция распределения c c=' + str(c)))
    plt.title('График функций распределения c различными параметрами c' )

plt.xlabel('x')
plt.ylabel('сdf(x)')
plt.grid()
plt.legend()
plt.show()
```

```python
from src.CharFuncInverter.Bohman import BohmansInverters as bi
```

```python
inv = bi.BohmanE()

c = 2
mu = 0

l = Levy(c, mu)

inv.fit(l.chr)

x = np.linspace(mu, 5, 1000) 
cdf_true = l.cdf(x) 

plt.plot(x, cdf_true, label=('Функция распределения с c=' + str(c)))

cdf_calc = inv.cdf(x)
cdf_calc = cdf_calc - cdf_calc[0]  ###получаем сдвинутую ф-ю распределения
plt.plot(x, cdf_calc, label=('Функция распределения аппроксимированная c c=' + str(c)))

plt.title('График функции распределения' )

plt.xlabel('x')
plt.ylabel('сdf(x)')
plt.grid()
plt.legend()
plt.show()
```
